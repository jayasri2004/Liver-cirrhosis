```python
import os
import numpy as np
import torch
import cv2
import matplotlib.pyplot as plt
from glob import glob
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from transformers import SegformerImageProcessor, SegformerForSemanticSegmentation
import torch.nn.functional as F
from tqdm.notebook import tqdm
import evaluate

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Enable synchronous CUDA debugging (optional, for debugging GPU issues)
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

# Paths to images and masks
image_paths = sorted(glob("/kaggle/input/cirrhosis-corrected/Cirrhosis_T2_2D/train/*/images/*.png"))
mask_paths = sorted(glob("/kaggle/input/cirrhosis-corrected/Cirrhosis_T2_2D/train/*/masks/*.png"))
print(f"Total Images: {len(image_paths)}, Total Masks: {len(mask_paths)}")

# Dataset class
class SemanticSegmentationDataset(Dataset):
    def __init__(self, image_paths, mask_paths, image_processor, num_classes=2):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.image_processor = image_processor
        self.num_classes = num_classes

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load grayscale image and convert to 3-channel (Segformer expects RGB)
        image = Image.open(self.image_paths[idx]).convert("L")
        image = np.array(image)
        image = np.stack([image] * 3, axis=-1).astype(np.uint8)  # Stack to create 3 channels
        image = Image.fromarray(image)

        # Load segmentation mask
        mask = Image.open(self.mask_paths[idx])
        mask = np.array(mask, dtype=np.int64)

        # Clean mask values (assuming masks have 0 and 255)
        mask[mask == 255] = 1  # Map 255 to 1 (foreground class)
        mask = np.clip(mask, 0, self.num_classes - 1)  # Ensure values are in [0, num_classes-1]

        # Apply image processor
        encoded_inputs = self.image_processor(image, mask, return_tensors="pt")
        for k, v in encoded_inputs.items():
            encoded_inputs[k].squeeze_()  # Remove batch dimension
        return encoded_inputs

# Image processor
image_processor = SegformerImageProcessor(do_resize=True, size=(512, 512), reduce_labels=False)

# Dataset and DataLoader
train_dataset = SemanticSegmentationDataset(image_paths, mask_paths, image_processor, num_classes=2)
train_dataloader = DataLoader(train_dataset, batch_size=5, shuffle=True)

# Model initialization
num_classes = 2
model = SegformerForSemanticSegmentation.from_pretrained(
    "nvidia/mit-b0",
    num_labels=num_classes,
    ignore_mismatched_sizes=True
).to(device)

# Custom CEDiceLoss definition
class CEDiceLoss(torch.nn.Module):
    def __init__(self, dice_weight=0.5, smooth=1e-6):
        super(CEDiceLoss, self).__init__()
        self.dice_weight = dice_weight  # Weight for Dice Loss vs Cross-Entropy Loss
        self.smooth = smooth  # Smoothing factor to avoid division by zero
        self.ce_loss = torch.nn.CrossEntropyLoss()  # Built-in Cross-Entropy Loss

    def dice_loss(self, logits, targets):
        num_classes = logits.size(1)
        dice = 0
        for c in range(num_classes):
            pred = logits[:, c].sigmoid()  # Apply sigmoid to get probabilities
            true = (targets == c).float()  # One-hot encoding for the class
            intersection = (pred * true).sum()
            union = pred.sum() + true.sum()
            dice += (2. * intersection + self.smooth) / (union + self.smooth)
        return 1 - dice / num_classes  # Return 1 - Dice coefficient as loss

    def forward(self, logits, targets):
        # Upsample logits to match target size
        logits = F.interpolate(logits, size=targets.shape[-2:], mode="bilinear", align_corners=False)
        ce_loss = self.ce_loss(logits, targets)  # Compute Cross-Entropy Loss
        dice_loss = self.dice_loss(logits, targets)  # Compute Dice Loss
        return (1 - self.dice_weight) * ce_loss + self.dice_weight * dice_loss  # Combine losses

# Initialize loss function
loss_fn = CEDiceLoss(dice_weight=0.5).to(device)

# Optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)

# Metric for evaluation
metric = evaluate.load("mean_iou")

# Training loop
model.train()
step_counter = 0
losses = []
mean_ious = []
steps = []

for epoch in range(1, 11):  # Train for 10 epochs
    print(f"\nEpoch {epoch}")
    epoch_losses = []
    
    for idx, batch in enumerate(tqdm(train_dataloader)):
        pixel_values = batch["pixel_values"].to(device)
        labels = batch["labels"].to(device)
        
        # Forward pass
        optimizer.zero_grad()
        outputs = model(pixel_values=pixel_values)
        logits = outputs.logits
        loss = loss_fn(logits, labels)
        loss.backward()
        optimizer.step()
        
        epoch_losses.append(loss.item())
        
        # Evaluate every 100 steps
        if idx % 100 == 0:
            with torch.no_grad():
                upsampled_logits = F.interpolate(
                    logits, size=labels.shape[-2:], mode="bilinear", align_corners=False
                )
                predicted = upsampled_logits.argmax(dim=1)
                metric.add_batch(
                    predictions=predicted.detach().cpu().numpy(),
                    references=labels.detach().cpu().numpy()
                )
                metrics = metric.compute(num_labels=num_classes, ignore_index=-100, reduce_labels=False)
                current_iou = metrics['mean_iou']
                
                # Store metrics for plotting
                losses.append(loss.item())
                mean_ious.append(current_iou)
                steps.append(step_counter)
                
                print(f"Step {step_counter} | Loss: {loss.item():.4f} | Mean IoU: {current_iou:.4f}")
        
        step_counter += 1
    
    avg_epoch_loss = sum(epoch_losses) / len(epoch_losses)
    print(f"Epoch {epoch} | Average Loss: {avg_epoch_loss:.4f}")

# Save model weights
torch.save(model.state_dict(), "segformer_model_weights.pth")

# Plot training metrics
plt.figure(figsize=(12, 5))

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(steps, losses, '-o', markersize=2, alpha=0.7)
plt.title('Training Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.grid(True, linestyle='--', alpha=0.6)

# Plot Mean IoU
plt.subplot(1, 2, 2)
plt.plot(steps, mean_ious, '-o', markersize=2, alpha=0.7, color='orange')
plt.title('Mean IoU')
plt.xlabel('Steps')
plt.ylabel('IoU')
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('training_metrics.png')
plt.show()
```

```python
# Inference function
def infer_segmentation(model, image_path, image_processor, device, num_classes=2):
    model.eval()  # Set model to eval mode

    # Load and preprocess image
    image = Image.open(image_path).convert("L")
    image_np = np.array(image)
    image_np = np.stack([image_np] * 3, axis=-1).astype(np.uint8)  # Convert grayscale to 3-channel
    image_rgb = Image.fromarray(image_np)

    # Preprocess using the processor
    inputs = image_processor(image_rgb, return_tensors="pt").to(device)

    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits  # shape: (1, num_classes, H, W)
        upsampled_logits = F.interpolate(
            logits, size=(image.size[1], image.size[0]), mode="bilinear", align_corners=False
        )
        predicted_mask = upsampled_logits.argmax(dim=1).squeeze().cpu().numpy()

    return predicted_mask

# Visualize prediction
def visualize_prediction(image_path, predicted_mask):
    original = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.imshow(original, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(predicted_mask, cmap='jet', interpolation='nearest')
    plt.title("Predicted Segmentation")
    plt.axis("off")

    plt.show()

# Example usage
test_image_path = image_paths[0]  # Pick any image path you want to infer on
predicted = infer_segmentation(model, test_image_path, image_processor, device)
visualize_prediction(test_image_path, predicted)
```
